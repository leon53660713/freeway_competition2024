# import package

```python
# webcrawler
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
# data
import pandas as pd
```

# use webdriver

## Firefox

```python
# webdriver
driver_path = "D:\geckodriver\geckodriver.exe"
service = Service(driver_path)
```

# get in page

```python
# open driver
## firefox
driver = webdriver.Firefox(service=service)
# ## chrome
# driver = webdriver.Chrome(service=service)

# page url
url = 'https://www.cwa.gov.tw/V8/C/D/DailyPrecipitation.html'
driver.get(url)

# click id : SID
driver.find_element(By.ID, "SID").click()
# choose index == 2
# index count by 0
select = Select(driver.find_element(By.ID, "SID"))
select.select_by_index(2)

# click id : Year
driver.find_element(By.ID, "Year").click()
# choose index == 1
# index count by 0
select = Select(driver.find_element(By.ID, "Year"))
select.select_by_index(1)

# wait 5 sec
driver.implicitly_wait(5)

# save table
table = WebDriverWait(driver, 5).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, "body > div.wrapper > main > div > div > div:nth-child(1)"))
)
# change to df
df = pd.read_html(table.get_attribute('outerHTML'))[0]
# tell finish
print("finish")

# quit
driver.quit()
```
